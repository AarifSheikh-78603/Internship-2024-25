# Exploring the Deals & Investments of Shark - Amit Jain

## Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

import plotly.graph_objects as go
import plotly.express as px  # Import plotly.express
```

## Import Dataset

```python
df = pd.read_csv('Shark Tank India.csv')
```

## Breakdown of Pitches, Present Days, and Investments season by season of Shark - Amit Jain

```python
# Filter out investments where the amount is greater than 0
df_invested = df[df['Amit Investment Amount'] > 0]

# Calculate the total number of pitches per season
pitches = df.groupby('Season').size()

# Calculate the total present days per season
present_days = df.groupby('Season')['Amit Present'].sum()

# Calculate the total number of investments per season (where investment amount > 0)
investments = df_invested.groupby('Season').size()

# Create a DataFrame with the required values
data = pd.DataFrame({
    'Total Pitches': pitches,
    'Total Present Days': present_days,
    'Total Investments': investments
})

# Plotting
fig, ax = plt.subplots(figsize=(8, 4))

# Plot the Total Pitches bar
pitches_bar = ax.bar(data.index, data['Total Pitches'], label='Total Pitches', color='#CAF0F8', edgecolor='black', linewidth=0.2)

# Plot the Total Present Days bar within the Total Pitches bar
present_days_bar = ax.bar(data.index, data['Total Present Days'], label='Total Present Days', color='#90E0EF', edgecolor='black', linewidth=0.2)

# Plot the Total Investments bar within the Total Present Days bar
investments_bar = ax.bar(data.index, data['Total Investments'], label='Offered Investments', color='#00B4D8', edgecolor='black', linewidth=0.2)

# Add labels and title
ax.set_xlabel('Season')
ax.set_ylabel('Pitches')
ax.set_title('Pitches, Attendance & Investments by Season of Shark - Amit Jain', color='r', fontsize=12, fontstyle='oblique')

# Setting yticks to integer values
ymax = 240
plt.yticks(np.arange(0, ymax + 1, step=40))  # Adjust the step value as needed

# Set x-ticks to integer values
ax.set_xticks(data.index)
ax.set_xticklabels(data.index)

plt.grid(linestyle='-', linewidth=0.3, axis='y')

# Show legend
ax.legend(loc='upper right')

# Adding actual values to each stack
for container in ax.containers:
    ax.bar_label(container, label_type='edge', fontstyle='oblique', fontsize=11, color='black')

# Display the plot
plt.show()

```

## Distribution of Deals Based-on Pitches (Season 1)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filter out investments where the amount is greater than 0
df_invested = df[df['Amit Investment Amount'] > 0]

# Calculate the total number of pitches per season
pitches = df.groupby('Season').size()

# Calculate the total present days per season
present_days = df.groupby('Season')['Pitch'].sum()

# Calculate the total number of investments per season (where investment amount > 0)
investments = df_invested.groupby('Season').size()

# Create a DataFrame with the required values
data = pd.DataFrame({
    'Total Pitches': pitches,
    'Total Present Days': present_days,
    'Total Investments': investments
})

# Filter data for Season 1
season_1_data = data.loc[1]  # Assuming Season 1 is labeled as 1

# Calculate the percentage of investments relative to present days
investment_percentage = (season_1_data['Total Investments'] / season_1_data['Total Pitches']) * 100
present_days_percentage = 100 - investment_percentage

# Prepare data for pie chart
labels = ['Successful Deal', 'Un-successful Deal']
sizes = [investment_percentage, present_days_percentage]
colors = ['#FAD02C', '#E9EAEC']
explode = (0.1, 0)  # Explode the first slice for emphasis

# Define a function to display both percentage and actual values
def func(pct, allvals):
    absolute = int(round(pct/100.*np.sum(allvals)))
    return "{:.1f}%\n({:d})".format(pct, absolute)

# Plotting the pie chart
fig, ax = plt.subplots(figsize=(4, 4))
ax.pie(sizes, explode=explode, labels=labels, colors=colors, autopct=lambda pct: func(pct, [season_1_data['Total Investments'], season_1_data['Total Pitches'] - season_1_data['Total Investments']]), startangle=90, shadow=True, counterclock=False)

# Add title
ax.set_title('Deal Success Rate By Pitches (Season 1)')

# Display the plot
plt.show()
```

## Gender Distribution (Season 1)

```python
# Filter data for Season 1 and Amit's investments greater than 0
season_1_data = df[df['Season'] == 1]
Amit_investments = season_1_data[season_1_data['Amit Investment Amount'] > 0]

# Sum the number of presenters by gender for these filtered investments
presenters_data = Amit_investments[['Male Presenters', 'Female Presenters', 'Transgender Presenters']].sum()

# Plotting the pie chart
plt.figure(figsize=(6, 6))  # Increased figure size for better visibility
wedges, _, autotexts = plt.pie(
    presenters_data,
    colors=['#ff6f61', '#88b04b', '#b565a7'],  # Customized color scheme
    autopct='%1.2f%%',
    pctdistance=0.75,  # Pulls the percentage labels closer to the center
    shadow=True,  # Adds a shadow for a 3D effect
    wedgeprops={'edgecolor': 'black', 'linewidth': 1, 'antialiased': True}  # Enhances the wedge borders
)

# Increase font size for the percentage values
for autotext in autotexts:
    autotext.set_fontsize(14)  # Set font size to 14

# Adding a center circle to create a donut chart look
centre_circle = plt.Circle((0, 0), 0.50, fc='white')
plt.gca().add_artist(centre_circle)

# Adding a legend within the white circle area
plt.legend(wedges, ['Male', 'Female', 'Transgender'], title="Gender", loc="center", fontsize=12)

# Customizing the title
plt.title('Gender Distribution (Season 1)',
          fontsize=16, color='red', fontweight='bold', loc='center')

# Adjusting the layout for better spacing
plt.tight_layout()

# Displaying the plot
plt.show()
```
